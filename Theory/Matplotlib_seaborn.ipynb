```python
import matplotlib.pyplot as plt

x1 = [2,3,4,5,6]
y1 = [10,20,30,40,50]

x2 = [1,3,5,7,9]
y2 = [5,15,25,35,45]

plt.scatter(x1,y1,label="x1y1_data")
plt.scatter(x2,y2,label="x2y2_data")
plt.legend()
plt.xlabel("X-values")
plt.ylabel("Y-values")
plt.title("Data_points")
plt.show()
```

```python
#Bar chart

fruits = ["Orange","Apple","Mango","Kiwi","Pears","Grapes"]
todays_count = [100,130,200,90,60,40]
yesterdays_count = [90,60,250,120,80,10]

plt.bar(fruits,todays_count,label="Today")
plt.bar(fruits,yesterdays_count,label="Yesterday")
plt.legend()
plt.title("Fruits count")
plt.xlabel("Fruits")
plt.ylabel("Count")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
x = np.linspace(1,10,100)

y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y1,label="sine_curve")
plt.plot(x,y2,label="cosine_curve")

plt.xlabel("Angle in degrees")
plt.ylabel("Values")
plt.legend()

plt.show()
```

```python
fig,ax1 = plt.subplots()
ax2 = ax1.twinx()

y2 = np.exp(x)

ax1.plot(x,y1,label="sine_curve")
ax2.plot(x,y2,label="exp_curve")

ax1.set_ylabel("sine_values")
ax2.set_ylabel("exponential_values")
ax1.set_xlabel("Angle in Degrees")

plt.title("sine & exp on different axes")
plt.legend()

plt.show()
```

```python
y2 = np.exp(x)

plt.plot(x,y1,label="sine_curve")
plt.plot(x,y2,label="exp_curve")

plt.show()
```

```python
#subplots using gridspec

from matplotlib import gridspec

fig = plt.figure(figsize=(12,8))
gs = gridspec.GridSpec(2,2) #(no of rows and no of cols)

fig1 = fig.add_subplot(gs[0,0])
fig2 = fig.add_subplot(gs[0,1])
fig3 = fig.add_subplot(gs[1,0])
fig4 = fig.add_subplot(gs[1,1])

fig1.plot(x,np.sin(x),label="sin_curve",color="g",linestyle="--",marker="o",linewidth=1)
fig2.plot(x,np.cos(x),label="cosine_curve",color="b",linestyle="-",marker="^",linewidth=1.2)
fig3.plot(x,np.exp(x),label="exponential_curve",color="y",linestyle="-.",marker=".",linewidth=1.4)
fig4.plot(x,np.log(x),label="logarthmic_curve",color="r",linestyle=":",marker=",",linewidth=1.6)

fig1.set_xlabel("Angle in Degrees")
fig1.set_ylabel("Sin Values")
fig1.set_title("Sin Curve")

fig2.set_xlabel("Angle in Degrees")
fig2.set_ylabel("Cosine Values")
fig2.set_title("Cosine Curve")

fig3.set_xlabel("x_val")
fig3.set_ylabel("exp Values")
fig3.set_title("Exponential Curve")

fig4.set_xlabel("x_val")
fig4.set_ylabel("Log Values")
fig4.set_title("Logarthmic Curve")

plt.tight_layout()
plt.show()
```

```python
#Direct option to create the subplots

x = np.linspace(1,20,100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(12,4))

plt.subplot(1,2,1) #(no of rows,no of cols,position)
plt.plot(x,y1,label="sine_curve")
plt.xlabel("Angle in degrees")
plt.ylabel("Sine values")

plt.subplot(1,2,2)
plt.plot(x,y2,label="cosine_curve")
plt.xlabel("Angle in degrees")
plt.ylabel("Cosine_values")
```
