Tasks:

Find the total and average salary for each department.

Find the highest salary in each department.

Add a new column that shows the average salary of each department and assign it to each employee.

Find the total salary of employees grouped by Experience range:

0-5 years: 'Junior'

6-10 years: 'Senior'

```python
import pandas as pd

data = {'Department': ['HR', 'IT', 'HR', 'IT', 'Finance', 'Finance', 'IT', 'HR'],
        'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Helen'],
        'Salary': [50000, 70000, 45000, 80000, 60000, 62000, 72000, 48000],
        'Experience': [5, 7, 4, 10, 8, 6, 9, 3]}
```

```python
df = pd.DataFrame(data)
df
```

```python
#Find the total and average salary for each department

df_avg_sal = df.groupby(by="Department",as_index=False)["Salary"].mean()
df_avg_sal
```

```python
#Find the highest salary in each department.

df.groupby(by="Department",as_index=False)["Salary"].max()
```

```python
#Add a new column that shows the average salary of each department and assign it to each employee

df_mod = pd.merge(df,df_avg_sal,how="left",on="Department")
df_mod
```

```python
df_mod.rename(columns={"Salary_x":"Salary",
                       "Salary_y":"Avg_Salary"},inplace=True)
df_mod
```

```python
#using apply function

def get_avg_salary(dept):
    df_avg_sal = df.groupby(by="Department",as_index=False)["Salary"].mean()
    avg_sal_val = df_avg_sal[df_avg_sal["Department"]==dept]["Salary"].iloc[0]
    return avg_sal_val

df["Avg_salary"] = df["Department"].apply(lambda val:get_avg_salary(val))

df
```

```python
"""Find the total salary of employees grouped by Experience range:
0-5 years: 'Junior'
6-10 years: 'Senior'"""

def get_category_name(exp_val):
    if exp_val<=5:
        return "Junior"
    elif exp_val>=6 and exp_val<=10:
        return "Senior"

df["employee_category"] = df["Experience"].apply(lambda val:get_category_name(val))

df
```

```python
df.groupby(by="employee_category",as_index=False)["Salary"].sum()
```
